```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv("../Dataset/Beginning Datasets/train.csv")
test_df = pd.read_csv("../Dataset/Beginning Datasets/test.csv")
```

## Relationship Between Variables

```python
# Correlation Matrix

corr = train_df.corr()

np.fill_diagonal(corr.values, 0)

corr.replace (0,np.nan,inplace=True)
plt.show()

corr
```

```python
#With SNS Heatmap

f,ax = plt.subplots(figsize = (15,8))
sns.heatmap(train_df.corr(), robust=True, annot=True, linewidths=1, fmt= '.2f',ax=ax,cmap= 'icefire')
plt.title('Train data heat map plot')
```

```python
# Heatmap plot for test dataset
f,ax = plt.subplots(figsize=(15,8))
sns.heatmap(test_df.corr(), robust=True, annot=True, linewidths=1, fmt= '.2f',ax=ax, cmap= 'icefire')
plt.title('Test data heat map plot');
```

```python
# Highest correlations between all of our variables

corr.unstack().sort_values(kind='quicksort', na_position='first').drop_duplicates(keep='last')
```

```python
# Display highest correlations between price_range and the other features in our dataset

corr.abs()['price_range'].sort_values(ascending=False)
```
