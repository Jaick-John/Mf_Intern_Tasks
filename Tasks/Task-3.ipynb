## Line Plot



```python
#Randomized Plotting From the Dataset

X = list(range(10))
Y = np.random.rand(10)
Z = np.random.rand(10)
plt.figure(figsize=[6,6])
plt.plot(X,Y,color='cyan',marker='x',linestyle='-.',linewidth=2,markersize=10)
plt.plot(X,Z,color='teal',marker='o',linestyle='-',linewidth=3,markersize=15)
plt.legend()
plt.show()
```

```python
#SeaBorn Clock Speed Line Plot

plt.figure(figsize=[14,9])
plt.subplot(3,1,1)
sns.lineplot(x=df.mobile_wt,y=df.clock_speed,data=df,size='dual_sim',sizes=['1','2.5'])
plt.subplot(3,1,2)
val=np.random.randint(0,len(color))
sns.lineplot(x=df.mobile_wt,y=df.clock_speed,data=df,hue='dual_sim',palette=color[val])
plt.suptitle("Seaborn LinePlot")
plt.subplot(3,1,3)
sns.lineplot(x=df.mobile_wt,y=df.clock_speed,data=df,style='dual_sim')
```

```python
#Different Variation Of Plots

plt.figure(figsize=[14,5])
plt.subplot(2,1,1)
sns.lineplot(x=df.mobile_wt,y=df.clock_speed,data=df,err_style='bars')
plt.subplot(2,1,2)
sns.lineplot(x=df.mobile_wt,y=df.clock_speed,data=df,err_style='band')
```

# Bar Plot

```python
#Using Matplotlib

X = df.n_cores.unique()
Y = df.groupby('n_cores')['battery_power'].median()
plt.figure(figsize=[14,5])
plt.bar(X,Y,width=0.5,color = '#FF7F7F',edgecolor = 'black',linewidth = 2, yerr = 150, ecolor = 'yellow')
```

```python
#Using Seaborn

plt.figure(figsize=[12,6])
val=np.random.randint(0,len(color))
sns.barplot(x='price_range',y='battery_power',data=df,order=[3,2,1,0],hue='dual_sim',hue_order=[1,0],errcolor='teal',estimator=median,palette=color[val])
#ordering the x axis 
```

# Pie Chart

```python
#Matplotlib

plt.figure(figsize=[6,6])
plt.pie(list(df.price_range.value_counts()),labels = df.price_range.unique())
plt.legend()
plt.show()
```

```python
x=np.random.randint(0,100,6)
plt.figure(figsize=[6,6])
plt.pie(x=x,explode=[0,0,0,0,0.15,0],shadow=True,labels=x)
plt.legend()
plt.show()
```
