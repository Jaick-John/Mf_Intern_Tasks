# Time Series Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Drop rows with null date_added
df.dropna(subset=['date_added'], inplace=True)

# Aggregate the number of titles added per month
df['date_added'] = df['date_added'].dt.to_period('M').dt.to_timestamp()
df_ts = df['date_added'].value_counts().sort_index().reset_index()
df_ts.columns = ['ds', 'y']

# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(df_ts['ds'], df_ts['y'], marker='o')
plt.xticks(rotation=90)
plt.title('Number of Titles Added Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Titles')
plt.show()

# Forecasting with Prophet
model = Prophet()
model.fit(df_ts)
future = model.make_future_dataframe(periods=12, freq='M')
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
plt.title('Forecast of Titles Added Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Titles')
plt.show()
```

## Clustering


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../../Datasets_Mainflow/disney_plus_titles.csv')

# Preprocess the 'duration' column
def preprocess_duration(duration):
    if 'min' in duration:
        return int(duration.replace(' min', ''))
    elif 'Season' in duration:
        return int(duration.split()[0]) * 60  # Assuming 1 season = 60 minutes (or any appropriate average)
    return None

df['duration'] = df['duration'].apply(preprocess_duration)

# Drop rows with null values in 'release_year' and 'duration'
numeric_features = df[['release_year', 'duration']].dropna()

# Standardize features
scaler = StandardScaler()
numeric_features_scaled = scaler.fit_transform(numeric_features)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(numeric_features_scaled)

# Add the cluster labels to the dataframe
df['cluster'] = -1
df.loc[numeric_features.index, 'cluster'] = clusters

# Create a dictionary to map cluster numbers to descriptions
cluster_labels = {0: 'Cluster 1: Early Short Movies', 1: 'Cluster 2: Recent Long movies', 2: 'Cluster 3: Recent Short movies'}

# Visualize the clusters with a legend
plt.figure(figsize=(10, 10))
scatter = plt.scatter(numeric_features_scaled[:, 0], numeric_features_scaled[:, 1], c=clusters, cmap='viridis')
plt.title('KMeans Clustering of Disney Plus Titles')
plt.xlabel('Standardized Release Year')
plt.ylabel('Standardized Duration')

# Create a legend with cluster descriptions
legend1 = plt.legend(handles=scatter.legend_elements()[0], labels=[cluster_labels[i] for i in range(3)], loc="best", title="Clusters")
plt.gca().add_artist(legend1)

plt.show()

# Print the first few rows of the dataframe with the cluster labels
print(df.head())
```
